```python
x1 = [-1, 0, 1, -1, 0, 1]
x2 = [1, 0.5, 1, -0.5, -1, -0.5]


def d1(x,y):
  return(((x-(-1))**2 + (y-1)**2)**(1/2))

def d2(x,y):
  return(((x-1)**2 + (y-(-0.5))**2)**(1/2))

for i in range(len(x1)):
  print(i+1)
  print(d1(x1[i],x2[i]))
  print(d2(x1[i],x2[i]))

```

```python
import matplotlib.pyplot as plt

centerx = [-1,1]
centery = [1,-0.5]

cluster1x = [-1,0,-1] ; cluster1y = [1,0.5,-0.5]
cluster2x = [1,0,1] ; cluster2y = [1,-1,-0.5]
plt.scatter(cluster1x,cluster1y,c='blue')
plt.scatter(cluster2x,cluster2y,c='pink')
plt.scatter(centerx, centery, c ='red',marker='x')
plt.legend(['cluster 1','cluster 2','center'])
plt.show()
```
